```python
# Dependencies
from matplotlib import pyplot as plt
import numpy
import pandas as pd
import statistics
```

```python
# Read CSV
city_data = "city_data.csv"
city_data_df = pd.read_csv(city_data)

city_data_df
```

```python
ride_data = "ride_data.csv"
ride_data_df = pd.read_csv(ride_data)

ride_data_df
```

```python
combined_city_df = pd.merge(city_data_df,ride_data_df, how='outer', on='city')
combined_city_df
```

```python
rides = [0,5,10,15,20,25,30,35,40]
type = ["Urban", "Suburban", "Rural"]
color = ["orange", "blue", "yellow"]

plt.scatter(numpy.arange(rides, type, color="orange", label="Urban"))
plt.scatter(numpy.arange(rides, type, color="blue", label="Suburban"))
plt.scatter(numpy.arange(rides, type, color="yellow", label="Rural"))

plt.legend(loc="best")

plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare")

plt.axis("equal")
plt.show()
```

```python
#Pie Label
labels = ["Rural", "Urban","Suburban"]

#Pie value
size = [185, 172, 100 , 110]

#pie Color
colors = ["pink", "yellow", "blue"]

explode = (0.1,0,0,0)
```

```python
plt.pie(size, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.show()
```
